```python
def solve_prioritizing_problem(issues, developers):
    linear_problem_objective = f'Maximize the number of critical tasks prioritized.'
    solver = pywraplp.Solver(linear_problem_objective, pywraplp.Solver.CBC_MIXED_INTEGER_PROGRAMMING)
    
    prioritize_matrix = [[0 for x in range(len(developers))] for y in range(len(issues))]
    
    # 1. Create the variables we want to optimize
    for issue_index, issue in issues.iterrows():
        for dev_index, dev in developers.iterrows():
            prioritize_matrix[issue_index][dev_index] = solver.IntVar(0, solver.infinity(), 
                                                           f'is_issue_{issue_index}_assigned_for_dev_{dev_index}?')
    # 2. Add constraints for each resource    

    # 2.1 Dev working hours limitation
    for dev_index, dev in developers.iterrows():
        used_hours_sum_expression = 0
        for issue_index, issue in issues.iterrows():
            used_hours_sum_expression += prioritize_matrix[issue_index][dev_index] * issue['effort']
        solver.Add(used_hours_sum_expression <= dev['working_hours'])
    
    # 2.2 Only one dev per issue limitation
    for issue_index, issue in issues.iterrows():
        allocated_issue_sum_expression = sum(prioritize_matrix[issue_index][:])
        solver.Add(allocated_issue_sum_expression <= 1)

    # 3. Maximize the objective function
    maximization_expression = 0
    for issue_index, issue in issues.iterrows():
        for dev_index, dev in developers.iterrows():
            maximization_expression += prioritize_matrix[issue_index][dev_index] * issue['criticality']
    solver.Maximize(maximization_expression)
    
    # Solve problem
    status = solver.Solve()
    
    # If an optimal solution has been found, print results
    if status == pywraplp.Solver.OPTIMAL:
        print('================= Solution =================')
        print(f'Solved in {solver.wall_time():.2f} milliseconds in {solver.iterations()} iterations')
        print()
        print(f'Optimal value (sum of prioritized tasks criticalities) = {solver.Objective().Value()}')
        print()
        for dev_index, dev in developers.iterrows():
            print('-----------------------------------------------------------------------------')
            print(f'üë®‚Äçüíª {dev["name"]} assignee in issues: ')
            print()
            sum_of_used_dev_hours = 0
            for issue_index, issue in issues.iterrows():
                if(prioritize_matrix[issue_index][dev_index].solution_value() == 0):
                    continue
                print(f"‚Ä¢ [criticality: {issue['criticality']}| effort: {issue['effort']}] {issue['title']}")
                sum_of_used_dev_hours += issue['effort']
            print()
            print(f'üïí {dev["name"]} hours used: {sum_of_used_dev_hours}/{dev["working_hours"]} | Remaining hours: {dev["working_hours"] - sum_of_used_dev_hours}')
            print('-----------------------------------------------------------------------------')
            print()
            
        print('-----------------------------------------------------------------------------')
        print(f'‚ùå Unassigned issues: ')
        print()
        for issue_index, issue in issues.iterrows():
            if(sum(list(map(lambda x: x.solution_value(), prioritize_matrix[issue_index][:]))) == 0):
                print(f"‚Ä¢ [criticality: {issue['criticality']}| effort: {issue['effort']}] {issue['title']}")
        print('-----------------------------------------------------------------------------')
    else:
        print('The solver could not find an optimal solution.')
```

```python
# Generating random issues

def generate_random_issues(number_of_issues):
    possibles_efforts = [1, 4, 8, 12, 20, 40]
    possibles_criticities = [1, 2, 5, 10]
    
    issues = pd.DataFrame()
    
    for index in range(0, number_of_issues):
        issues = pd.concat(
            [issues, pd.DataFrame(
                [{
                    'title': issue_titles_mock[random.randint(0, len(issue_titles_mock)-1)], 
                    'effort': possibles_efforts[random.randint(0, len(possibles_efforts)-1)], 
                    'criticality': possibles_criticities[random.randint(0, len(possibles_criticities)-1)]
                }]
            )],
            ignore_index=True
        )

    return issues
```

```python
# Generating random devs

def generate_random_devs(number_of_devs, number_of_weeks = 1):
    possibles_working_hours_per_week = [20, 40]
    
    developers = pd.DataFrame()
    
    for index in range(0, number_of_devs):
        developers = pd.concat(
            [developers, pd.DataFrame(
                [{
                    'name': dev_names_mock[random.randint(0, len(dev_names_mock)-1)], 
                    'working_hours_per_week': possibles_working_hours_per_week[random.randint(0, len(possibles_working_hours_per_week)-1)]
                }]
            )],
            ignore_index=True
        )
        
    developers['working_hours'] = developers['working_hours_per_week'] * number_of_weeks
    return developers
```
